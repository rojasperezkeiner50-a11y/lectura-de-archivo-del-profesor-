# TRATAMIENTO DE DATOS — SESIÓN 1
Actividad práctica de exploración, limpieza y transformación.

```python
import pandas as pd
import numpy as np

# Cargar dataset
df = pd.read_csv('ventas_clientes.csv')
df.head()
```

## Exploración Inicial

```python
df.info()
df.describe(include='all')
```

## Identificación de Problemas

```python
df.isna().sum()
df.duplicated().sum()
df['ciudad'].unique()
```

## Limpieza de Datos

```python
df['ciudad'] = df['ciudad'].str.lower().str.strip().str.capitalize()
df['fecha_venta'] = pd.to_datetime(df['fecha_venta'], errors='coerce')
df = df.drop_duplicates()
df['unidades'] = df['unidades'].fillna(1)
df['nombre'] = df['nombre'].fillna("No registrado")
df['precio'] = df['precio'].abs()
df['total'] = df['unidades'] * df['precio']
df = df[df['correo'].str.contains("@", na=False)]
df.head()
```

## Exportar Dataset Limpio

```python
df.to_csv('ventas_limpias.csv', index=False)
```
